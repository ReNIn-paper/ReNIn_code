```python
import h5py
import matplotlib.pyplot as plt
def check_dataset(filename, idx=0):
    with h5py.File(filename,'r') as f:
        print(f.keys())
        f_num_list = list(f.keys())
        # f_number_list = list(map(parse,f_num_list))
        f_number_list = list(range(len(f_num_list)))
        print("f_number_list",f_number_list)
        for i,key in enumerate(f.keys()):
            print(key,f[key].shape)
            plt.subplot(1,len(f.keys()),f_number_list[i]+1)
            plt.title(f"{key}\n{idx}")
            plt.imshow(f[key][idx])
            plt.axis('off')
        plt.pause(0.01)
        
```

```python
filename = data_list[0]
print(filename)
with h5py.File(filename,'r') as f:
    for i,key in enumerate(f.keys()):
        length = len(f[key])
print("length:",length)
interval = length // (length//100)
if interval == 0:
    interval = 1
print("interval",interval)
# check_dataset(filename=filename,idx=0)
for idx in range(0,length,interval):
    check_dataset(filename,idx)
    plt.pause(0.01)
```
